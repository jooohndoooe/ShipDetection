## Exploratory data analysis of the dataset ##

First we import the libraries requiered for the data analysis, import the csv file containing ship segmentations and declare folder paths.

Next we take a sample from the train_ship_segmentations and iterate that sample

Each iteration we print out the data in the train_ship_segmentation and the corresponding image from the test_v2 folder

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
from PIL import Image


train_ship_segmentations = pd.read_csv('data/train_ship_segmentations_v2.csv')

train_folder = 'data/train_v2/'
test_folder = 'data/test_v2/'

train_data_sample = train_ship_segmentations.sample(3)

for data_entry in train_data_sample.iterrows():
    print(data_entry[1])

    img = mpimg.imread(train_folder + data_entry[1]['ImageId'])

    plt.imshow(img)
    plt.show()
```

Having grasped the strucutre of the data next we analyze the images contained in the test_v2 and train_v2 folders

We iterate through images in the train and test folder add store their sizes and print out the set of unique values

To speed up the process insted of iterating through all images, we take a sample of 1000 from each folder

Lastly we print out the number of images in each folder

```python
image_sizes = []

for filename in random.sample(os.listdir(train_folder), 1000):
    image =  Image.open(os.path.join(train_folder, filename))
    image_sizes.append(image.size)
for filename in random.sample(os.listdir(test_folder), 1000):
    image =  Image.open(os.path.join(test_folder, filename))
    image_sizes.append(image.size)

print(set(image_sizes))

print(len(os.listdir(train_folder)))
print(len(os.listdir(test_folder)))
```

After analyzing the imgaes in the train and test folders we see that they and uniformal, each being 768x768 pixels each.

Test folder contains 192556 images, Train folder contains 15606 images.
